```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import load_model
```

```python
word_indexes = imdb.get_word_index()

reverse_dict = dict()

for key, value in word_indexes.items():
    reverse_dict[value] = key
```

```python
model = load_model('../artifacts/simplernn.keras')
model.summary()
```

```python
model.get_weights()
```

```python
def preprocess(text:str):
    words = text.lower().split()
    encoded_word = [word_indexes.get(word, 2) + 3 for word in words]
    padded_word = pad_sequences([encoded_word], 500)

    return padded_word

def predict(text):
    preprocessed_input = preprocess(text)

    return model.predict(preprocessed_input)
```

```python
sample = 'This movie was really bad! The acting was horrible and the plot was really bad.'

predict(sample)
```
