```python
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import Embedding, Input
from tensorflow.keras.utils import pad_sequences # Future me, this implements a "max" length of the sentence
from tensorflow.keras.models import Sequential
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.optimizers import Adam

import numpy as np
```

```python
text = ["The glass of milk",
        'The glass of juice',
        'The cup of tea',
        'I am a good boy',
        'Does he know',
        'Our brains are shrinking',
        'tomorrow is the promise of today']
```

```python
# Create the vocab
voc_size = 10_000

ohe_repr = []

for sentence in text:
    ohe_repr.append(one_hot(sentence, voc_size))

ohe_repr
```

```python
sent_length = 8
embedded_docs = pad_sequences(ohe_repr,
                              padding = 'pre',
                              maxlen = sent_length)

embedded_docs
```

```python
#Training the model
dim = 10

model = Sequential()
model.add(Input((sent_length, )))
model.add(Embedding(voc_size, dim))

optimizer = Adam(learning_rate = 0.01)
loss = MeanSquaredError()

model.compile(optimizer = optimizer, loss = loss)

model.summary()
```

```python
model.predict(embedded_docs)
```
